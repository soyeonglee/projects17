## 선물 이론가 산출

```python
def futures_price(S,r,d,T,t0):
    t = np.busday_count(t0.date(), T.date())/252
    F = S * np.exp((r-d)*d)
    return F
```

```python
T= pd.to_datetime('2017-12-14')
r=0.0225
d=0.015
for n in (df.index):
    t0 = n
    S = df.loc[n,'지수']
    F = futures_price(S,r,d,T,t0)
    f='{:,.2f}'.format(F)
    df.loc[n,'이론가1']=f
```

## 보급형 공식을 이용한 선물 이론가 계산

```python
def S_futures_price(S,r,d,T,t0):
    t = (T-t0).days / 365
    F = S * (1+(r-d)*t)
    return F
```

```python
T=pd.to_datetime('2017-12-14')
r=0.0225
d=0.015
for n in (df.index):
    t0 = n
    S = df.loc[n,'지수']
    F=S_futures_price(S,r,d,T,t0)
    f='{:,.2f}'.format(F)
    df.loc[n,'이론가2']=f
```

# 11.2 WTI 선물가격을 이용한 롤오버 비용 
## 엑셀에서 WTI 현물, 최근월물, 차근월물 데이터 읽어오기

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_excel('C:/Users/ldm98/Desktop/WTI.xlsx', 'Spot', index_col='Date')
df2 = pd.read_excel('C:/Users/ldm98/Desktop/WTI.xlsx', 'Front', index_col='Date')
df3 = pd.read_excel('C:/Users/ldm98/Desktop/WTI.xlsx', 'Second', index_col='Date')
```

## 시세 데이터프레임에 최종거래일 표시하기

```python
df['Expiry'] = np.nan
df.head()
```

```python
for d in df.index:
    if d in pd.to_datetime(expiry.values):
        df.loc[d].Expiry=1
df
```

```python
price_now=float(df['2017-12-28']['Spot'])
price_before= float(df['2010-01-04']['Spot'])
profit = (price_now - price_before)*1000
print('손익(현재가격- 초기가격) :', '{:,.2f}'.format(profit))
```

## 손익 계산: 선물에 투자했을 경우

```python
roll_over_cost_acc = 0
for d in df.index:
    if(df.loc[d].Expiry==1):
        roll_over_cost = df.loc[d].Second - df.loc[d].Front
        roll_over_cost_acc = roll_over_cost_acc + roll_over_cost
        
        print('누적 롤오버 비용: ', '{:,.2f}'.format(roll_over_cost_acc))
```

```python
price_now = float(df['2017-12-28']['Front'])
price_before = float(df['2010-01-04']['Front'])
profit = (price_now - price_before - roll_over_cost_acc) *1000
print('누적 손익', '{:,.2f}'.format(profit))
```
